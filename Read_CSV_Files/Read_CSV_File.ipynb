```
#Create Own Schema

from pyspark.sql.types import StructType, StructField, IntegerType, StringType
schema_defined = StructType([StructField('Year', IntegerType(), True),
                             StructField('Name', StringType(),True),
                             StructField('County', StringType(),True),
                             StructField('Sex', StringType(), True),
                             StructField('Count', IntegerType(),True)
                            ])
```

```
df= spark.read.format("csv").schema(schema_defined).option("header", True ).option("sep",",").load("/FileStore/tables")
display (df)
```

```
schema_alternate = 'Year INTEGER, Name STRING, County STRING, Sex STRING, Count INTEGER'
```

```
df = spark.read.format("csv").schema(schema_alternate).option("header", True ).option("sep",",").load("/FileStore/tables")

display(df)
```
