```
#Create sample Dataframe
data_student = [("Raja", "Science",88,"p",90),
                ("Rakesh","Maths", 90,"p",70),
                ("Rama", "English", 20,"F",80),
                ("Ramesh","Science",45,"F",75),
                ("Rajesh","Maths", 30,"F",50),
                ("Raghav","Maths", None, "NA",79)
               ]

Schema= ["name","Subject","Mark","Status","Attendance"]

df = spark.createDataFrame(data = data_student, schema = Schema)

display (df)
```

```
#Update the existing column

from pyspark.sql.functions import when

df1 = df.withColumn("Status", when(df.Mark >=50,"Pass")
                    .when (df.Mark < 50,"Fail")
                    .otherwise ("Absentee"))

display (df1)
```

```
#Create a New Column
from pyspark.sql.functions import when

df2 = df.withColumn("New_status",when (df.Mark >50,"Pass")
                    .when(df.Mark <58, "Fail")
                    .otherwise("Absentee"))

display (df2)
```

```
#Another Syntax method
from pyspark.sql.functions import expr

df3=df.withColumn("new_status", expr ("CASE WHEN Mark >= 50 THEN 'Pass'" +
                                   "WHEN Mark < 50 THEN 'Fail'"+
                                   "ELSE 'Absentee' END"))

display (df3)
```

```
#Multi-conditions using AND and OR Operators
from pyspark.sql.functions import when

df4 = df.withColumn("Grade", when((df.Mark >=89) & (df.Attendance >=88), "Distinction")
                    .when((df.Mark >=50) & (df.Attendance >= 50), "Good")
                    .otherwise("Average"))

display(df4)
```

```
from pyspark.sql.functions import when

df4 = df.withColumn("Grade", when((df.Mark >=89) | (df.Attendance >=88), "Distinction")
                    .when((df.Mark >=50) & (df.Attendance >= 50), "Good")
                    .otherwise("Average"))

display(df4)
```
