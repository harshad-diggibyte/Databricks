```
#Create Employee Dataframe
employee_data = [(10, "Raj Kumar","1999","180","M", 2000),
                 (20, "Rahul Rajan", "2002","200","F",8000),
                 (39, "Raghav","2919","198","",6800),
                 (40, "Raja Singh","2004","100","F",7000),
                 (50, "Rama Krish", "2008","400","M", 1000),
                 (60,"Rasul","2014","500","M", 5000),
                 (78, "Kumar Chand","2984","680","M", 5000)
                ]

employee_schema= ["employee_1d","name","doj","employee_dept_id","gender","salary"]

employeeDF = spark.createDataFrame(data=employee_data, schema= employee_schema)

display(employeeDF)
```

```
#Create Department Dataframe

department_data = [("HR",100),("Supply",200),("Sales",300),("Stock",400)]

department_schema = ["dept_name","dept_id"]

departmentDF = spark.createDataFrame(data=department_data,schema = department_schema)

display(departmentDF)
```

```
display(employeeDF)

display(departmentDF)
```

```
#Inner Join

df_join = employeeDF.join(departmentDF,employeeDF.employee_dept_id == departmentDF.dept_id,"inner")

display(df_join)
```

```
#Full Outer Join
df_join = employeeDF.join(departmentDF,employeeDF.employee_dept_id == departmentDF.dept_id,"outer")

display(df_join)
```

```
#Left Outer Join

df_join = employeeDF.join(departmentDF,employeeDF.employee_dept_id == departmentDF.dept_id,"left")

display(df_join)
```

```
#Right Outer Join

df_join = employeeDF.join(departmentDF,employeeDF.employee_dept_id == departmentDF.dept_id,"right")

display(df_join)
```

```
#Left Semi Join
df_join = employeeDF.join(departmentDF,employeeDF.employee_dept_id == departmentDF.dept_id,"leftsemi")

display(df_join)
```

```
#Left anti join
df_join = employeeDF.join(departmentDF,employeeDF.employee_dept_id == departmentDF.dept_id,"leftanti")

display(df_join)
```
