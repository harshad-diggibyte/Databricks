```
#Create Sample Dataframe
empDF = [('John',[4, 6, 7, 9, 2]),
         ('David',[5, 1, 4, 7,1]),
         ('Mike', [6, 9, 1, 8, 2])
        ]

df = spark.createDataFrame(data=empDF, schema = ['Name','Array_col'])
df.show()
```

```

from pyspark.sql import functions as F

df_sort = df.withColumn("Sorted",F.array_sort(df["Array_col"]))
df_sort.show()
```
