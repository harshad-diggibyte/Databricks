```
#Create DataFrame

employee_data = [(10, "Raj", "Kumar","1999","180","M", 2000),
                 (20, "Rahul","Rajan", "2002","200","F",8000),
                 (39, "Raghav","Bales","2919","198",None,6800),
                 (40, "Raja","Singh","2004","100","F",7000),
                 (50, "Rama","Krish", "2008","400","M", 1000),
                 (60,"Rasul","Mishra","2014","500","M", 5000),
                 (78, "Kumar","Chand","2984","680","M", 5000)
                ]

employee_schema= ["employee_1d","First_name","Last_name","doj","employee_dept_id","gender","salary"]

empDF = spark.createDataFrame(data=employee_data, schema= employee_schema)

display(empDF)

```

```
#Add new column using constant literal

from pyspark.sql.functions import lit

empDF_AddColumn = empDF.withColumn("Location", lit("Mumbai")).show()
```

```
#Add new column by calculation

from pyspark.sql.functions import concat

empDF_AddColumn = empDF.withColumn("Bonus",empDF.salary*0.1)

display(empDF_AddColumn)
```

```
empDF_AddColumn = empDF.withColumn("Bonus",empDF.salary*0.1).withColumn("Name",concat("First_name" , "Last_name")).show()
```

```
empDF_AddColumn = empDF.withColumn("Bonus",empDF.salary*0.1).withColumn("Name",concat("First_name",lit(" "),"Last_name")).show()
```

```
#Rename a Column

empDF_RenameColumn = empDF_AddColumn.withColumnRenamed("Name","Full_Name").show()
```

```
#Drop a column

empDF_DropColumn= empDF_AddColumn.drop("Name").show()
```
