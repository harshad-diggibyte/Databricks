```
#Row Count Per Partition
#Create Sample Dataframe for Baby Names data

df = spark.read.format("csv").option("inferSchema",True).option("header",True).option("sep",",").load("/FileStore/tables/babynames.csv")
display(df)                   
```

```
#Number of Records in the Dataframe
print(df.count())
```

```
#Default Partition Count
print(df.rdd.getNumPartitions())
```

```
#Number of Records per Partition
from pyspark.sql.functions import spark_partition_id
df.withColumn("partitionId",spark_partition_id()).groupBy("partitionId").count().show()
```

```
#Repartition the Dataframe to 5
df_5 =df.select(df.year,df.name,df.percent,df.sex).repartition(5)
```

```
#Get Number of Partitions in the Dataframe
print(df_5.rdd.getNumPartitions())
```

```
#Get Number of Records per Partition
df_5.withColumn("partitionId", spark_partition_id()).groupBy("partitionid").count().show()
```
